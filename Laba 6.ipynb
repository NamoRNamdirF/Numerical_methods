```python
from pylab import *
from scipy.linalg import *

x = np.array[0,10]
y = np.array[0,10]

# задаем вектор m = [x**2, x, E]
m=vstack((x**2,x,ones(11))).T
# находим коэффициенты при составляющих вектора m
s=lstsq(m,y)[0]

# на отрезке [-5,5]
x_prec=linspace(-5,5,101)
# рисуем теоретическую кривую x<sup>2</sup>
plot(x_prec,x_prec**2,'--',lw=2)
# рисуем точки
plot(x,y,'D')
# рисуем кривую вида y = ax<sup>2</sup> + bx + c, подставляя из решения коэффициенты s[0], s[1], s[2]
plot(x_prec,s[0]*x_prec**2+s[1]*x_prec+s[2],'-',lw=2)

grid()
legend(('Initial','Data','Least square fit'),9)
savefig('plot4.png')
```

```python
%matplotlib notebook 
x = np.linspace(0, 5, 100)                     # от -5 до 2 сделать 100 точек
fig, ax = plt.subplots()                        # будет 1 график, на нем:
ax.plot(x, a +b/ x, color="blue", label="f'(x)")      # функция y1(x), синий, надпись y(x)
ax.plot(x,a*exp(b*x), color="red", label="f''(x)")      # функция y2(x), красный, надпись y'(x)
ax.set_xlabel("x")                              # подпись у горизонтальной оси х
ax.set_ylabel("y")                              # подпись у вертикальной оси y
ax.legend()                                     # показывать условные обозначения
plt.grid(which = 'major',color = 'k')
plt.show()                                      # показать рисунок
```
